```python
class Expenses:
    monthly_expenses = {}

    def calculate_expenses(month):
        inventory = int(input("Enter inventory expense: "))
        salary = int(input("Enter staff salary expense: "))
        other = int(input("Enter other expenses: "))

        Expenses.monthly_expenses[month] = {
            "inventory": inventory,
            "salary": salary,
            "other": other
        }

        total = inventory + salary + other
        return total
```

```python
class Revenue:
    revenue = 0

    def calculate_profit(total_expenses):
        Revenue.revenue = int(input("Enter montly revenue: "))
        profit = Revenue.revenue - total_expenses

        print("\n----- Monthly Report -----")
        print("Total Revenue: Rs", Revenue.revenue)
        print("Total Expenses: Rs", total_expenses)
        print("Profit: Rs", profit)       

        
```
